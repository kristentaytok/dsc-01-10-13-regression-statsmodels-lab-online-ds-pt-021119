### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
#R-squared of the TV vs. sales is 0.612, which suggests this model is a better fit than the baseline model: 61% of variations in sales is explained by spend on TV channels present in this model. There is a slight negative skew and kurtosis is ~2.8, which is much lower than 3*standard deviation. The Prob(F-statistic)/P-value is < 0.05, which suggests this sample will consistently represent the population.
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
df_predict = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(df_predict)
prediction_model = model.predict(df_predict)
print(prediction_model)
# make predictions for those x values and store them


# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='TV', y='sales')
plt.plot(df_predict, prediction_model)
plt.show()
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on residuals
#The positive relationship on the "Y and Fitted vs. X" shows that TV spend and sales are positively correlated. The "Residuals vs. height" graph shows a slight pattern in the distance between the dots and the mean value (regression line) increasing across the regression line, which suggests slight heteroscedasticity 
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
# For all the variables, check if they hold normality assumption
f = 'sales~radio'
model = ols(formula=f, data=df).fit()
df_predict2 = pd.DataFrame({'Radio': [df.radio.min(), df.radio.max()]})
print(df_predict2)

prediction_model2 = model.predict(df_predict2)
print(prediction_model2)
df.plot(kind='scatter', x='radio', y='sales')
plt.plot(df_predict2, prediction_model2)
plt.show()
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Radio", fig=fig)
plt.show()
```
